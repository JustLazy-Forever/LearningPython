```python
TOLERABLE_ERROR = 10**-5

def func(x):
    return x**2 - 5 * x + 5
```

```python
def print_iteration_table(rows) -> None:
    print("\nBy\t\t:- Darshana Ghodasaini")
    print("Roll no.\t:- KCE081BCT011")
    print("\nIteration Table:\n")
    print(f"{'Iterations':<11}{'a':>12}{'b':>12}{'m':>12}{'f(a)':>12}{'f(m)':>12}{'Error':>12}")

    for row in rows:
        print(
            f"{row['Iteration']:<11}"
            f"{row['a']:>12.6f}"
            f"{row['b']:>12.6f}"
            f"{row['m']:>12.6f}"
            f"{row['f(a)']:>12.6f}"
            f"{row['f(m)']:>12.6f}"
            f"{row['Error']:>12.6f}"
        )
```

```python
def findRoot(a, b) -> list:
    iterations = 0
    rows = []
    while True:
        m = (a + b) / 2
        func_a = func(a)
        func_m = func(m)
        error = abs(b - a) / 2

        rows.append({
            "Iteration": iterations + 1,
            "a": a,
            "b": b,
            "m": m,
            "f(a)": func_a,
            "f(m)": func_m,
            "Error": error,
        })

        if func_m * func_a < 0:
            b = m
        else:
            a = m

        iterations += 1
        if abs(func_m) < TOLERABLE_ERROR:
            return [m, iterations, rows]
```

```python
# Example run (non-interactive)
# Choose initial guesses a and b such that func(a)*func(b) < 0
a = 1.0
b = 2.0
root, iterations, rows = findRoot(a, b)
print_iteration_table(rows)
print()
print(f"Root: {root:.6f}")
print(f"Iterations: {iterations}")
```
